```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
Flags = [cv2.IMREAD_COLOR, cv2.IMREAD_GRAYSCALE, cv2.IMREAD_UNCHANGED ]

flag = Flags[1]

I_XY = cv2.imread('o105i186_8.tif', flag)


plt.imshow(I_XY, cmap='gray', vmin=0, vmax=255)
plt.colorbar()
plt.show()

y,x = 100,270
#---------------
pix_value = I_XY[y,x]

print("el tamaño de la imagen es de ", np.size(I_XY[0,:]), " x ",np.size(I_XY[:,0]), " píxeles.")
```

```python
#recortar región de la imagen
#pixeles de inicio en la esquina superior izquierda en y vs x
Y_sup,X_sup = 100 , 350
#píxeles de final en la esquina derecha inferior
Y_inf,X_inf = 150, 500

I_XY_recortado=I_XY[Y_sup:Y_inf,X_sup:X_inf]
```

```python
plt.imshow(I_XY_recortado, cmap='gray', vmin=0, vmax=255)
plt.colorbar()
plt.show()
```
